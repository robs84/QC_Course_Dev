Introduction to superposition and entanglement will highlight two frequently used gates, the Hadamard and CNOT (or CX) gates.  The Hadamard gate takes a qubit of known state and yields a normalized linear combination of states in that basis.  The CNOT gate will perform a NOT operation on the second qubit if the first (control) qubit is $|1\rangle$ and will not perform any operation if the control qubit is $|0\rangle$, in the computational basis.

A useful gate:
$CX = |00\rangle\langle00| + |01\rangle\langle01| + |10\rangle\langle01| + |01\rangle\langle10|$

```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

In this module we will look at both superposition and entanglement using the Hadamard and CNOT gates.
The learning objective is to gain an intuition about these fundamental operations.

```python
#the quantum register we will be operating on and the classical register to measure with
qr = QuantumRegister(2)
cr = ClassicalRegister(2)

#the circuit that is composed of both the quantum and classical register
circ1 = QuantumCircuit(qr,cr)

'''
Adding a Hadamard gate to the circuit. Recall that the Hadamard tapplied to a single qubit 
of a given state outputs a normalized linear superposition of states in that 
(computational or +/-) basis.
'''
circ1.h(qr[0])
circ1.draw(output = 'mpl')
```

```python
#a CNOT gate entangles the two qubits in our quantum register
circ1.cx(qr[0],qr[1])
circ1.measure(qr,cr)
simulator = Aer.get_backend('qasm_simulator')
result1 = execute(circ1, backend = simulator).result()
circ1.draw(output = 'mpl')
```

```python
plot_histogram(result1.get_counts(circ1)) #approximately 50/50
```

Now, repeat the process with a second Hadamard on the first qubit before the CNOT and 
again after the CNOT.  Comment on what happens.  This logic can be worked out somewhat 
trivially using Dirac notation.

```python
circuit2 = QuantumCircuit(qr,cr)
circuit2.h(qr[0])
circuit2.h(qr[0])
circuit2.cx(qr[0],qr[1])
circuit2.measure(qr,cr)
result2 = execute(circuit2, backend = simulator).result()
circuit2.draw(output = 'mpl')
```

```python
plot_histogram(result2.get_counts(circuit2)) #100% in the |00> state
```

```python
circuit3 = QuantumCircuit(qr,cr)
circuit3.h(qr[0])
circuit3.cx(qr[0],qr[1])
circuit3.h(qr[0])
circuit3.measure(qr,cr)
result3 = execute(circuit3, backend = simulator).result()
circuit3.draw(output = 'mpl')
```

```python
plot_histogram(result3.get_counts(circuit3))#approximately evenly divided among the four states
```
